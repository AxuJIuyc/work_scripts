```python
def comparison(img1, img2, hash_size=8):
    hash_1 = imagehash.average_hash(img1, hash_size)
    hash_2 = imagehash.average_hash(img2, hash_size)
    print(f"average_hash_{hash_size}:", hash_1 - hash_2)

    hash_1 = imagehash.phash(img1, hash_size)
    hash_2 = imagehash.phash(img2, hash_size)
    print(f"phash_{hash_size}:", hash_1 - hash_2)

    hash_1 = imagehash.phash_simple(img1, hash_size)
    hash_2 = imagehash.phash_simple(img2, hash_size)
    print(f"phash_simple_{hash_size}:", hash_1 - hash_2)

    hash_1 = imagehash.dhash(img1, hash_size)
    hash_2 = imagehash.dhash(img2, hash_size)
    print(f"dhash_{hash_size}:", hash_1 - hash_2)

    hash_1 = imagehash.dhash_vertical(img1, hash_size)
    hash_2 = imagehash.dhash_vertical(img2, hash_size)
    print(f"dhash_vertical_{hash_size}:", hash_1 - hash_2)

    hash_1 = imagehash.whash(img1, hash_size)
    hash_2 = imagehash.whash(img2, hash_size)
    print(f"whash_{hash_size}:", hash_1 - hash_2)

    hash_1 = imagehash.colorhash(img1)
    hash_2 = imagehash.colorhash(img2)
    print("colorhash:", hash_1 - hash_2)


    plt.figure()
    f, axarr = plt.subplots(1,2) 
    axarr[0].imshow(img1)
    axarr[1].imshow(img2)
```

```python
img1 = Image.open('2023-03-06.15-01-52.814.png')
img2 = Image.open('2023-03-06.15-05-24.738.png')
comparison(img1, img2, 128)
```

```python
import json

def search_box(json_path):
    with open(json_path, 'r') as f:
        json_obj = json.load(f)
    json_shapes = json_obj['shapes']
    for shape in json_shapes:
        if shape['label'] == 'box':
            return shape['points']
```

```python
def crops(name1, name2, hash_size=8):

    json1 = f'{name1}.json'
    json2 = f'{name2}.json'
    box1 = search_box(json1)
    box2 = search_box(json2)
    print(box1)
    print(box2)

    img1 = Image.open(f'{name1}.png')
    img2 = Image.open(f'{name2}.png')

    crop1 = img1.crop((box1[0][0], box1[0][1], box1[1][0], box1[1][1]))
    crop2 = img2.crop((box2[0][0], box2[0][1], box2[1][0], box2[1][1]))

    plt.figure()
    f, axarr = plt.subplots(2,2) 
    axarr[0][0].imshow(img1)
    axarr[1][0].imshow(img2)
    axarr[0][1].imshow(crop1)
    axarr[1][1].imshow(crop2)
    
    comparison(crop1, crop2, hash_size)
```

```python
comparison(crop1, crop2, 32)
```

```python

name1 = '2023-03-02.17-00-24.712'
name2 = '2023-03-02.16-55-20.677'
crops(name1, name2, 32)
```
