```python
import os
import json
# import ijson
from matplotlib import pyplot as plt
```

```python
def train_results(work_dir, log_file):

    with open(f"{work_dir}/{log_file}", 'r') as f:
        logs = f.readlines()

    epoch, it, hm_loss, acc_pose, loss, PCK, NME, mAP = 0, 0, 0, 0, 0, 0, 0, 0
    results = []

    for step in logs[1:]:
        j = json.loads(step)
        if j['mode'] == 'train':
            epoch = j['epoch']
            it = j['iter']
            hm_loss = j['heatmap_loss']
            acc_pose = j['acc_pose']
            loss = j['loss']
        elif j['mode'] == 'val':
            if 'PCK' in j: 
                PCK = j['PCK']
            if 'NME' in j: 
                NME = j['NME']
            if 'mAP' in j: 
                mAP = j['mAP']
        results.append({"epoch": epoch, "iter": it, "heatmap_loss": hm_loss, 
                        "acc_pose": acc_pose, "loss": loss, "PCK": PCK, 
                        "NME": NME, "mAP": mAP})
    return results

```

```python
def plot_show(name, results, modules=['hml', 'accp', 'loss', 'PCK', 'NME', 'mAP']):
    x, y1, y2, y3, y4, y5, y6 = [], [], [], [], [], [], [], 

    for label, i in zip(results, range(len(results))):
        x.append(i)
        y1.append(label['heatmap_loss'])
        y2.append(label['acc_pose'])
        y3.append(label['loss'])
        y4.append(label['PCK'])
        y5.append(label['NME'])
        y6.append(label['mAP'])

    if 'hml' in modules:
        plt.plot(x, y1, label='heatmap_loss')
    if 'accp' in modules:
        plt.plot(x, y2, label='acc_pose')
    if 'loss' in modules:
        plt.plot(x, y3, label='loss')
    if 'PCK' in modules:
        plt.plot(x, y4, label='PCK')
    if 'NME' in modules:
        plt.plot(x, y5, label='NME')
    if 'mAP' in modules:
        plt.plot(x, y6, label='mAP')

    plt.xlabel('steps')
    plt.ylabel('metrics')
    plt.title(name)
    plt.legend()

    plt.show()
```

```python
work_dir = "../work_dirs/pprofile_3"
print(os.listdir(work_dir))
log_file = "20230330_143301.log.json"
out_file = f"{work_dir}/res_{log_file}"

results = train_results(work_dir, log_file)
with open(out_file, 'w') as f:
    json.dump(results, f)
```

```python
plot_show('pprofile', results, ['accp', 'PCK'])
```
